```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('hydro_august_2020_10_339.jpg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur = cv2.blur(img,(500,500))
```

```python
plt.imshow(img)
```

```python
plt.imshow(blur)
```

```python
import numpy as np
```

```python
mask = np.zeros((1080,1920,3), dtype=np.uint8)
mask = cv2.rectangle(mask, (1570, 0),(1890, 150), (255,255,255), -1)


out = np.where(mask==(0,0,0), img, blur)
```

```python
plt.imshow(out)
```
